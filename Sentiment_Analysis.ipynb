```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/Tweets_sentiments.csv')
```

```python
df
```

```python
del df['Unnamed: 0']
```

```python
df.dtypes
```

```python
df['TimeStamps'] = pd.to_datetime(df['TimeStamps'])
```

```python
df.dtypes
```

```python
df = df.sort_values(by='TimeStamps', ascending=True)
```

```python
df = df.reset_index()
```

```python
df
```

```python
del df['index']
```

```python
df
```

```python
df['Sentiment'] = df['Sentiment'].replace({'Positive': 1, 'Negative': -1})
```

```python
df
```

```python
df_senti = df.groupby(['TimeStamps']).Sentiment.sum()
df_senti
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df_senti.index, df_senti, marker='o', linestyle='-', color='b', label='Sentiment')
plt.xlabel('Date')
plt.ylabel('Sentiment')
plt.title('Sentiment Analysis of ITC Demerger')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
```
